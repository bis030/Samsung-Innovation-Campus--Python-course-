```python
a=[[1,2,3],[5,6],[9]]
b= a.copy()
b[0].append(6)
a
```

```python
# to avoid this referencing problem in copy
# to avoid a from being changed by other list's operations

from copy import deepcopy

a=[[1,2,3],[5,6],[9]]
b= deepcopy(a)
b[0].append(6)
b
```
