```python
# find whether pt lies on which quadrant


x,y= map(int,input("enter y coordinate").split())  #take two inputs in single line


if x==0 and y==0 :
    print(f"({x},{y}) lie on origin")
elif x==0:
    print(f"({x},{y}) lie on y axis")
elif y==0:
    print(f"({x},{y}) lie on x axis")
elif x>0 and y>0:
    print(f"({x},{y}) lie on first quad")
elif x<0 and y>0:
    print(f"({x},{y}) lie on sec quad")
elif x<0 and y<0:
    print(f"({x},{y}) lie on third quad")
else:
    print(f"({x},{y}) lie on fourth quad")

```

```python

# print untill you get Y
b="abcxyz"
for x in b:
  if x == 'y':
    break
  print(x)


# print skip when you get Y
b="abcxyz"
for x in b:
  if x == 'y':
    continue
  print(x)
```
